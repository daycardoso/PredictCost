```python

# ======================
# 4. 5×2-cv Paired t-test (Otimizado para Memória)
# ======================
def five_two_cv_ttest_gpu(model1_cls, model2_cls, X, y, random_state=42):
    diffs = []
    for i in range(5):
        kf = KFold(n_splits=2, shuffle=True, random_state=random_state+i)
        for tr_idx, te_idx in kf.split(X):
            X_tr, X_te = X.iloc[tr_idx], X.iloc[te_idx]
            y_tr, y_te = y.iloc[tr_idx], y.iloc[te_idx]

            # OTIMIZAÇÃO: Treine e avalie um modelo de cada vez

            # Modelo 1
            m1 = model1_cls()
            m1.fit(X_tr, y_tr)
            # mean_squared_error from cuml.metrics returns a scalar
            e1 = mean_squared_error(y_te, m1.predict(X_te))
            del m1 # Libera modelo 1
            gc.collect()
            cp.get_default_memory_pool().free_all_blocks()

            # Modelo 2
            m2 = model2_cls()
            m2.fit(X_tr, y_tr)
            # mean_squared_error from cuml.metrics returns a scalar
            e2 = mean_squared_error(y_te, m2.predict(X_te))
            del m2 # Libera modelo 2
            gc.collect()
            cp.get_default_memory_pool().free_all_blocks()

            # Append the float difference directly
            diffs.append(e1 - e2)

            del X_tr, X_te, y_tr, y_te
            gc.collect()

    d = np.array(diffs).reshape(5, 2)
    d1, d2 = d[:, 0], d[:, 1]
    d_bar = d1.mean()
    S_sq = ((d1 - d2)**2).sum() / 5
    se = np.sqrt(S_sq)
    t_stat = d_bar / se
    p_val = 2 * (1 - t.cdf(abs(t_stat), df=5))
    return t_stat, p_val

rf_cls  = lambda: cuRF(**chosen_rf, random_state=42, n_streams=1)
xgb_cls = lambda: xgb.XGBRegressor(tree_method="gpu_hist", random_state=42, verbosity=0, **chosen_xgb)

t_stat, p_val = five_two_cv_ttest_gpu(rf_cls, xgb_cls, X_cudf, y_cudf)
print(f"\n5×2-cv t-test: t={t_stat:.4f}, p={p_val:.4f}")
```
